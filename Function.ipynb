```python
#Function without argument and no return type

def call():
    l=int(input("Enter l= "))#local variable
    b=int(input("Enter b= "))#local variable
    h=int(input("Enter h= "))#local variable
    a=l*b  #local variable
    v=l*b*h  #local variable
    print(a)
    print(v)
call()
   
```

```python
#Global Variable
l=int(input("Enter l= "))   #Global Variable
b=int(input("Enter b= "))#Global variable
h=int(input("Enter h= "))#Global variable
def call():
    global a,v #to make local --> Global
    a=l*b  #local variable
    v=l*b*h  #local variable
    print(a)
    print(v)

call()
```

```python
#Function with argument and no return type
def area(l,b):
    a=l*b
    print("Area=", a)
    
def vol(l,b,h):
    v=l*b*h
    print("Volume=",v)
    
l=int(input("Enter l= "))#global variable
b=int(input("Enter b= "))#global variable
h=int(input("Enter h= "))#global variable

area(l,b)
vol(l,b,h)
```

```python
def cal():
    l=int(input("Enter l= "))
    b=int(input("Enter b= "))
    h=int(input("Enter h= "))
    a=l*b
    v=l*b*h
    return a,v
x=cal()
print(x)
area,volume=x
print('The area= ',area)
print('The volume= ',volume)

```

```python
#Function with argument and return type
def call(l,b,h):
    area=l*b
    vol=l*b*h
    return area,vol
l=int(input("Enter l= "))
b=int(input('Enter b= '))
h=int(input("Enter h= "))
x=call(l,b,h)
area,volume=x
print('Area= ',area)
print('Volume= ',volume)
```

```python
#WAP to remove ',' from the following list and make it int: price=['10,000','20,300','1,00000','1,19,000']
price=['10,000','20,300','1,00000','1,19,000']
b=list(map(lambda x: x.replace(',',''),price))
a=list(map(lambda x: int(x),b))

print(a)
```
